```python
import cv2
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def display_img(img):
    fig = plt.figure(figsize=(15,15))
    ax = fig.add_subplot(111)
    ax.imshow(img, cmap='gray')
```

```python
img = cv2.imread('images/exame1-1.png')
```

```python
display_img(img)
```

```python
img2 = cv2.bilateralFilter(img, 9,75,75)
display_img(img2)
```

```python
kernel_5 = np.ones((5,5), dtype=np.float32)
kernel_3 = np.ones((3,3), dtype=np.float32)
kernel_1 = np.ones((1,1), dtype=np.float32)
```

```python
opening = cv2.morphologyEx(img, cv2.MORPH_OPEN, kernel_3)
display_img(opening)
```
