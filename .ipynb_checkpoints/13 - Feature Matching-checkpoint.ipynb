```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
def display_img(img):
    fig = plt.figure(figsize=(15,15))
    ax = fig.add_subplot(111)
    ax.imshow(img, cmap='gray')
```

# Brute Force - ORB

```python
reeses = cv2.imread('images/reeses_puffs.png', 0)
display_img(reeses)
```

```python
cereals = cv2.imread('images/many_cereals.jpg', 0)
display_img(cereals)
```

```python
orb = cv2.ORB_create()
```

```python
kp1, des1 = orb.detectAndCompute(reeses, None)
kp2, des2 = orb.detectAndCompute(cereals, None)
```

```python
bf = cv2.BFMatcher(cv2.NORM_HAMMING, crossCheck=True)
```

```python
matches = bf.match(des1, des2)
```

```python
matches = sorted(matches, key=lambda x:x.distance)
```

```python
len(matches)
```

```python
reeses_matches = cv2.drawMatches(reeses,kp1, cereals, kp2, matches[:25], None, flags=2)
```

```python
display_img(reeses_matches)
```

# Brute Force - SIFT

```python
sift = cv2.xfeatures2d.SIFT_create()
```

```python
kp1, des1 = sift.detectAndCompute(reeses, None)
kp2, des2 = sift.detectAndCompute(cereals, None)
```

```python
bf = cv2.BFMatcher()
```

```python
matches = bf.knnMatch(des1, des2, k=2)
```

```python
# Ratio test
# Less distance == better match
good = []
for match1, match2 in matches:
    if match1.distance < 0.75 * match2.distance:
        good.append([match1])
```

```python
print(len(good))
print(len(matches))
```

```python
sift_matches = cv2.drawMatchesKnn(reeses, kp1, cereals, kp2, good, None, flags=2)
display_img(sift_matches)
```

### Resume

```python
sift = cv2.xfeatures2d.SIFT_create()
kp1, des1 = sift.detectAndCompute(reeses, None)
kp2, des2 = sift.detectAndCompute(cereals, None)
bf = cv2.BFMatcher()
matches = bf.knnMatch(des1, des2, k=2)
# Ratio test
# Less distance == better match
good = []
for match1, match2 in matches:
    if match1.distance < 0.75 * match2.distance:
        good.append([match1])
sift_matches = cv2.drawMatchesKnn(reeses, kp1, cereals, kp2, good, None, flags=2)
```

# FLANN - Sift

```python
sift = cv2.xfeatures2d.SIFT_create()
kp1, des1 = sift.detectAndCompute(reeses, None)
kp2, des2 = sift.detectAndCompute(cereals, None)
```

```python
FLANN_INDEX_KDTREE = 0
index_params = dict(algorithm=FLANN_INDEX_KDTREE, trees=5)
search_params = dict(checks=50)
```

```python
flann = cv2.FlannBasedMatcher(index_params, search_params)
```

```python
matches = flann.knnMatch(des1, des2, k=2)
```

```python
matchesMask = [[0,0] for i in range(len(matches))]
```

```python
# Ratio test
# Less distance == better match
good = []
for i, (match1, match2) in enumerate(matches):
    if match1.distance < 0.7 * match2.distance:
        matchesMask[i] = [1,0]
```

```python
draw_parms = dict(matchColor=(0,255,0),
                 singlePointColor=(255,0,0),
                 matchesMask=matchesMask,
                 flags=0)
```

```python
flann_matches = cv2.drawMatchesKnn(reeses, kp1, cereals, kp2, matches, None, **draw_parms)
display_img(flann_matches)
```
