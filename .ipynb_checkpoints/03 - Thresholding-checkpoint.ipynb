```python
import cv2
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('images/exame1-1.png', 0)
```

```python
type(img)
```

```python
plt.imshow(img, cmap='gray')
```

```python
img.max()
```

```python
def show_pic(img):
    fig = plt.figure(figsize=(15,15))
    ax = fig.add_subplot(111)
    ax.imshow(img, cmap='gray')
```

```python
ret, img_threshold = cv2.threshold(img, 170, 255, cv2.THRESH_BINARY)
```

```python
show_pic(img_threshold)
```

```python
img_threshold2 = cv2.adaptiveThreshold(img, 255, cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY, 7, 8)
```

```python
show_pic(img_threshold2)
```

```python
blended = cv2.addWeighted(src1=img, alpha=0.5, src2=img_threshold2, beta=0.5, gamma=0)
show_pic(blended)
```
