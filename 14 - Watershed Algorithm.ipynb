```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
def display_img(img):
    fig = plt.figure(figsize=(15,15))
    ax = fig.add_subplot(111)
    ax.imshow(img, cmap='gray')
```

```python
sep_coins = cv2.imread('images/pennies.jpg')
display_img(sep_coins)
```

```python
sep_blur = cv2.medianBlur(sep_coins, 25)
```

```python
display_img(sep_blur)
```

```python
gray_sep = cv2.cvtColor(sep_blur, cv2.COLOR_BGR2GRAY)
display_img(gray_sep)
```

```python
ret, sep_th = cv2.threshold(gray_sep, 160, 255, cv2.THRESH_BINARY_INV)
display_img(sep_th)
```

```python
contours, hierarchy = cv2.findContours(sep_th.copy(), cv2.RETR_CCOMP, cv2.CHAIN_APPROX_SIMPLE)
```

```python
for i in range(len(contours)):
    if hierarchy[0][i][3] == -1:
        cv2.drawContours(sep_coins, contours, i, (255,0,0), 10)
display_img(sep_coins)
```
